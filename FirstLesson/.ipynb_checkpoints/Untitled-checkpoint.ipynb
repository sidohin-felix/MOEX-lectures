```python
#Now let's implement a printing of this data so we can use it somehow...
class dataEntry(object):
    
    age = None
    job = None
    martial = None
    education = None
    default = None
    balance = None
    housing = None
    loan = None
    contact = None
    day = None
    month = None
    duration = None
    campaign = None
    pdays = None
    previous = None
    poutcome = None
    y = None
    
    def __init__(self,data):
        self.age,self.job,self.martial,self.education,self.default,self.balance,\
        self.housing,self.loan,self.contact,self.day,self.month,self.duration,\
        self.campaign,self.pdays,self.previous,self.poutcome,self.y = data.split(';')
        return None
    
    def isSingle(self):
        if self.martial != '\"single\"':
            return False
        else:
            return True

    def __str__(self):
        return "age=" + self.age + ",job=" + self.job + ",single=" + self.martial
    
class dataEntryArray:
    
    entries = []
    
    def __init__(self,data):
        for element in data:
            self.entries.append(dataEntry(element))
```

```python
#1. берем временой ряд количества акций фонда F(t)
#2. вычисляем временной ряд разницы количества акций фонда, dF(t)
#3. берем временной ряд цен индекса Мосбиржы M(t)
#4. вычисляем временной ряд разницы цен индекса Мосбиржы, dM(t)
#5. Считаем количество случаев n, когда выполняется dF(t) > 0 и dM(t+1) > 0
#6. Вычисляем отношение (вероятность) P(A) = n/N , где N - это количество исследуемых дней.
```

```python
import random

F = []

for i in range(0,365):
    F.append(random.randint(180, 240))

print(F)
```

```python
F.reverse()
```

```python
print(F)
```

```python
y = F.pop()
dF = []

while len(F) > 0:
    dF.append(F.pop() - y)
    y = F.pop()

print(dF)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(0,len(dF)), dF, linewidth=1.0,color='g')
plt.show()
```

```python
print(len(F))
print(len(dF))
```

```python
import time

def readFileToArray(filename):
    f = open(filename,'r') # 'r' = read.
    data = f.readlines()
    f.close()
    return data

r = readFileToArray('datasets/bank-full.csv')

s = time.time()

entries = []
for element in r:
    entries.append(element.split(';'))

e = time.time()
print(e - s)
```
